```python
import pandas as pd
```

```python
hcmst = pd.read_csv("hcmst2017.csv", encoding="mac_roman")
```

```python
cols = ["Q5", "Q21A_Year", "Q21A_Month", "Q21B_Year", "Q21B_Month", "Q21C_Year", "Q21C_Month", "Q21D_Year", "Q21D_Month",]
colnames = ["same_sex", "year_met", "month_met", "year_rel", "month_rel", "year_movein", "month_movein", "year_marry", "month_marry"]
relevant = hcmst[cols]
relevant.columns = colnames
relevant = relevant.loc[relevant["year_met"].notna()]
relevant.head()
```

```python
clean = pd.DataFrame()
clean["same_sex"] = relevant["same_sex"].apply(lambda x: str(x).startswith("No"))
clean
```
